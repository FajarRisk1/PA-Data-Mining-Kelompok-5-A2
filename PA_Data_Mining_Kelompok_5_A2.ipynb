```python
def null_ratio(df1, limit=-1):
  percent_missing = df1.isnull().sum() * 100 / len(df1)
  df1_result = pd.DataFrame({
      "filled":df1.notnull().sum(),
      "N null":df1.isna().sum(),
      "null_ratio(%)": percent_missing,
      }
  )
  nr =  df1_result.sort_values("null_ratio(%)",ascending=False)
  return nr[nr["null_ratio(%)"] > limit]
```

```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv('dfsp_clean.csv')

# Split dataset into train and test set
train_data = data.sample(frac=0.8, random_state=42)
test_data = data.drop(train_data.index)

# Separate features and target variable
features_train = train_data.drop(['MEDV', 'MEDV_kategori'], axis=1)
features_test = test_data.drop(['MEDV', 'MEDV_kategori'], axis=1)
target_train = train_data['MEDV_kategori']
target_test = test_data['MEDV_kategori']

# Calculate the number of categories in the target variable
categories = target_train.unique()

# Calculate the probability of each category
priors = {}
for cat in categories:
    priors[cat] = len(target_train[target_train == cat]) / len(target_train)

# Calculate mean and standard deviation for each feature for each category
means = {}
stds = {}
for cat in categories:
    means[cat] = features_train[target_train == cat].mean()
    stds[cat] = features_train[target_train == cat].std()

# Define a function to calculate the probability density function of a feature
def gaussian_pdf(x, mean, std):
    exponent = np.exp(-((x - mean)**2 / (2 * std**2)))
    return (1 / (np.sqrt(2 * np.pi) * std)) * exponent

# Define a function to predict the category of a sample
def predict(sample):
    probabilities = {}
    for cat in categories:
        probabilities[cat] = priors[cat]
        for i in range(len(sample)):
            probabilities[cat] *= gaussian_pdf(sample[i], means[cat][i], stds[cat][i])
    return max(probabilities, key=probabilities.get)

# Predict the categories of the test set and calculate accuracy
correct = 0
for i in range(len(features_test)):
    sample = features_test.iloc[i]
    true_category = target_test.iloc[i]
    predicted_category = predict(sample)
    if true_category == predicted_category:
        correct += 1

accuracy_nb = correct / len(features_test)
print("Accuracy: {:.2f}%".format(accuracy_nb * 100))
```

```python
import pandas as pd
import numpy as np
import math

df = dfcl

df_encoded = pd.get_dummies(df, columns=['CRIM', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'LSTAT', 'MEDV'])
X = df_encoded.drop('MEDV_kategori', axis=1).values
y = df_encoded['MEDV_kategori'].values

X = np.nan_to_num(X, nan=np.nanmedian(X, axis=0))
y = pd.Series(y).fillna(pd.Series(y).mode()[0]).values

np.random.seed(42)
indices = np.random.permutation(len(X))
train_size = int(0.8 * len(X))
X_train = X[indices[:train_size]]
y_train = y[indices[:train_size]]
X_test = X[indices[train_size:]]
y_test = y[indices[train_size:]]

def euclidean_distance(x1, x2):
  distance = 0
  for i in range(len(x1)):
      distance += (x1[i] - x2[i]) ** 2
      return math.sqrt(distance)

def predict(X_train, y_train, x_test, k):
  distances = []
  for i in range(len(X_train)):
      distance = euclidean_distance(X_train[i], x_test)
      distances.append((distance, y_train[i]))
  distances.sort()
  neighbors = [distances[i][1] for i in range(k)]
  return max(set(neighbors), key=neighbors.count)
```

```python
y_pred = []
for x_test in X_test:
      y_pred.append(predict(X_train, y_train, x_test, k=5))
```

Class untuk memudahkan proses visualiasi,normalisasi, dan pengukuran cluster pada tiap algoritma cluster yang berbeda

```python
# Creating a class to handle clustering
class Cluster:
    
    # Pada constructor terjadi beberapa proses
    # 1. Membuat normalisasi data
    
    def __init__(self,train):
        self.train = train
        self.scaled = StandardScaler().fit_transform(self.train)
        self.clusters = None
        self.name = None
        self._label = None

    # Method apply adalah fungsi kedua yang harus dijalankan
    # Method apply akan menerima argumen tambahan sebagai argumen untuk 
    # algoritma cluster yang akan digunakan
    def apply(self,cluster_algo,**kwargs):
        agg_cluster = cluster_algo(**kwargs)


        self.clusters = agg_cluster.fit(self.scaled) #Set Value
        self._label = self.clusters.labels_ #Set value

        # Mengambil nama algoritma clusternya
        self.name = type(self.clusters).__name__

        return self
    
    # Df untuk membuat dataframe
    def df(self,normalized=False):

        if normalized:
            X = pd.DataFrame(self.scaled,columns=self.train.columns)
        else:
            X = self.train.copy(deep=True)
        
        X["Cluster"] = self._label
        X["Cluster"] = X['Cluster'].astype(object)

        return X
    # Untuk Mendapatkan cluster ke N
    def get_cluster(self,n,normalized=False):
        df = self.df(normalized)
        
        return df[df["Cluster"] == n]

    # Plot untuk membuat scatterplot 
    def plot(self,X,Y,normalized=False):
        return px.scatter(self.df(normalized), x=X, y=Y, color="Cluster")

    # Plot 3d untuk membuat scatterplot 3d
    def plot_3d(self,X,Y,Z,normalized=False):
        return px.scatter_3d(self.df(normalized), x=X, y=Y,z=Z, color="Cluster")

    # menggunakan decorator property untuk mengonversi method menjadi property
    @property
    def metric(self): #Method ini berfungsi untuk mengambil hasil pengukuran berdasrkan metric ...
        label = self._label
        # Ini metric yang digunakan 'silhouette_score', 'calinski_harabasz_score', 'davies_bouldin_score'
        
        return [silhouette_score(self.scaled,label),davies_bouldin_score(self.scaled,label),calinski_harabasz_score(self.scaled,label)]
```
